```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r'C:\Users\Gabi\Desktop\fashion-mnist_train.csv')
df
```

```python
X_train = df.drop('label',axis=1)
y_train = df['label']
```

```python
s = df['label']
s.value_counts()
```

```python
df['pixel10'].value_counts()
```

```python
pca = PCA()
X2D = pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
```

```python
d
```

```python
from sklearn.decomposition import IncrementalPCA
n_batches = 100
inc_pca = IncrementalPCA(n_components=187)
for X_batch in np.array_split(X_train, n_batches):
    inc_pca.partial_fit(X_batch)

X_reduced = inc_pca.transform(X_train)
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf4 = SVC()
clf5 = DecisionTreeClassifier()
voting_clf = VotingClassifier(
 estimators=[('lr', clf1), ('rf', clf2), ('sv', clf4),('dt', clf5)],
 voting='hard')
voting_clf.fit(X_reduced, y_train)
```

```python
dfRes=pd.read_csv(r'C:\Users\Gabi\Desktop\fashion-mnist_test.csv')
dfRes
```

```python
X_test = dfRes.drop('label',axis=1)
y_test = dfRes['label']
```

```python
pca1 = PCA(n_components=187)
X_reduced1 = pca1.fit_transform(X_test)
```

```python
voting_clf.predict(X_reduced1)
print(voting_clf.score(X_reduced1,y_test))
```
