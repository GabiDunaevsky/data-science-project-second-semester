```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
df=pd.read_csv(r'C:\Users\Gabi\Desktop\fashion-mnist_train.csv')
df
```

```python
X_train = df.drop('label',axis=1)
y_train = df['label']
```

```python
pca = PCA()
XDR = pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
```

```python
from sklearn.decomposition import IncrementalPCA
n_batches = 40
inc_pca = IncrementalPCA(n_components=187)
for X_batch in np.array_split(X_train, n_batches):
    inc_pca.partial_fit(X_batch)

X_train = inc_pca.transform(X_train)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
xgb_reg = xgb.XGBClassifier(eval_metric = 'mlogloss')
xgb_reg.fit(X_train, y_train)
```

```python
ad =AdaBoostClassifier(n_estimators=50, learning_rate=1.0, algorithm='SAMME.R')
```

```python
clf77= BaggingClassifier(n_estimators=10, random_state=0).fit(X_train, y_train)
```

```python
dt = BaggingClassifier(base_estimator=DecisionTreeClassifier(random_state=0),n_estimators=10, random_state=0).fit(X_train, y_train)
```

```python
ranfor = BaggingClassifier(base_estimator=RandomForestClassifier(),
                      n_estimators=10, random_state=0).fit(X_train, y_train)
```

```python
dfRes=pd.read_csv(r'C:\Users\Gabi\Desktop\fashion-mnist_test.csv')
dfRes
```

```python
X_test = dfRes.drop('label',axis=1)
y_test = dfRes['label']
```

```python
X_test = inc_pca.transform(X_test)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
xgb_reg.score(X_test, y_test)
```

```python
estimators = [('xg', xgb_reg ), ('cf', clf77),('ad', ad)]
```

```python
from sklearn.metrics import accuracy_score
```

```python
voting_clf = VotingClassifier(
 estimators=[('xg', xgb_reg ), ('cf', clf77),('ad', ad),('rn',ranfor),('dt',dt)],
 voting='soft')
voting_clf.fit(X_train, y_train)
y_pred = voting_clf.predict(X_test)
print(accuracy_score(y_pred, y_test))
```

```python
lala = StackingClassifier(
   estimators=estimators, final_estimator=ranfor,n_jobs=-1)
```

```python
lala.fit(X_train, y_train).score(X_test, y_test)
```
